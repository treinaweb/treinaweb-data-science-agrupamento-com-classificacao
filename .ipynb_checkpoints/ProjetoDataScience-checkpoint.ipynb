```python
def histograma_boxplot(recurso):
    fig, ax = plt.subplots(1, 2)
    ax[0].hist(recurso)
    ax[1].boxplot(recurso)
```

```python
histograma_boxplot(df['Finishing'])
```

```python
df_descricao = df.describe()
df_descricao
```

```python
histograma_boxplot(df_descricao.loc['min'])
```

# Tratamento Altura

```python
df['Height'] = df['Height'].str.split('\'')
df['Height']
```

```python
df['Height'] = [30.48 * int(elem[0]) + 2.54 * int(elem[1]) for elem in df['Height']]
df['Height']
```

```python
df
```

```python
histograma_boxplot(df['Height'])
```

# Tratamento Peso Jogador

```python
df['Weight'] = df['Weight'].str.split('l')
df['Weight']
```

```python
df['Weight'] = [int(elem[0]) * 0.453 for elem in df['Weight']]
df['Weight']
```

```python
df
```
